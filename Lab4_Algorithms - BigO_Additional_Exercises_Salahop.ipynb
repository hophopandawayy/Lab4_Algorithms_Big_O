```python
import time

# --- Setup ---
student_id = "TUPM-25-4175"
target = int(student_id[-2:])   # last two digits = 75
main_list = list(range(1, 1001))  # 1 to 1000
duplicate_list = [1, 2, 3, 4, 2, 5]

# --- Algorithms ---
def linear_search(arr, target):
    start = time.time()
    for i in range(len(arr)):
        if arr[i] == target:
            end = time.time()
            return i, end - start
    end = time.time()
    return -1, end - start

def binary_search(arr, target):
    start = time.time()
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            end = time.time()
            return mid, end - start
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    end = time.time()
    return -1, end - start

def find_duplicate(arr):
    start = time.time()
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                end = time.time()
                return arr[i], end - start
    end = time.time()
    return None, end - start

# --- Run and Compare ---
lin_index, lin_time = linear_search(main_list, target)
bin_index, bin_time = binary_search(main_list, target)
dup_value, dup_time = find_duplicate(duplicate_list)

print("Algorithm         | Result        | Time (seconds)")
print("-"*50)
print(f"Linear Search     | Index {lin_index:<5} | {lin_time:.6f}")
print(f"Binary Search     | Index {bin_index:<5} | {bin_time:.6f}")
print(f"Duplicate Finder  | Value {dup_value:<5} | {dup_time:.6f}")

# --- Dynamic Analysis ---
print("\nAnalysis:")
if lin_time > bin_time:
    print(f"- Binary search was faster than linear search for target {target}.")
else:
    print(f"- Linear search was surprisingly faster than binary search for target {target} (likely due to small input size).")

print(f"- Duplicate finder identified value {dup_value}, but took {dup_time:.6f} seconds with nested loops.")
print("- This shows how quadratic algorithms scale poorly compared to linear and logarithmic ones.")
```

```python
#Recursive vs. Iterative Strategy Analysis

import time

# --- Setup ---
student_id = "TUPM-25-4175"
n = int(student_id[-1]) + 15   # last digit (5) + 15 = 20
print("Problem size (n):", n)

# --- Recursive Fibonacci ---
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

start = time.time()
rec_result = fib_recursive(n)
end = time.time()
rec_time = end - start

# --- Iterative Fibonacci ---
def fib_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

start = time.time()
iter_result = fib_iterative(n)
end = time.time()
iter_time = end - start

# --- Comparative Table ---
print("\nApproach     | Result        | Time (seconds)")
print("-"*45)
print(f"Recursive    | {rec_result:<12} | {rec_time:.6f}")
print(f"Iterative    | {iter_result:<12} | {iter_time:.6f}")

# --- Dynamic Analysis ---
print("\nAnalysis:")
if rec_time > iter_time:
    print(f"- Iterative Fibonacci was faster than recursive for n={n}.")
else:
    print(f"- Recursive Fibonacci was unexpectedly faster for n={n} (likely due to small input size).")

print(f"- Recursive approach produced {rec_result}, but required repeated calls.")
print(f"- Iterative approach produced {iter_result}, with smoother execution.")
print("- Both are O(n), but recursion suffers from overhead compared to iteration.")
```

```python
#Sorting Algorithm Visual Comparator
import random, time

# --- Setup ---
student_id = "TUPM-25-4175"
list_size = int(student_id[-3:]) + 50   # last three digits (175) + 50 = 225
data = random.sample(range(1, list_size*2), list_size)  # random unique integers

# --- Sorting Algorithms ---
def bubble_sort(arr):
    comparisons, swaps = 0, 0
    start = time.time()
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            comparisons += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps += 1
    end = time.time()
    return comparisons, swaps, end - start

def selection_sort(arr):
    comparisons, swaps = 0, 0
    start = time.time()
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    end = time.time()
    return comparisons, swaps, end - start

def insertion_sort(arr):
    comparisons, swaps = 0, 0
    start = time.time()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            comparisons += 1
            arr[j+1] = arr[j]
            swaps += 1
            j -= 1
        arr[j+1] = key
    end = time.time()
    return comparisons, swaps, end - start

# --- Run and Compare ---
b_comp, b_swaps, b_time = bubble_sort(data[:])
s_comp, s_swaps, s_time = selection_sort(data[:])
i_comp, i_swaps, i_time = insertion_sort(data[:])

print("Algorithm       | Comparisons | Swaps | Time (seconds)")
print("-"*55)
print(f"Bubble Sort     | {b_comp:<11} | {b_swaps:<5} | {b_time:.6f}")
print(f"Selection Sort  | {s_comp:<11} | {s_swaps:<5} | {s_time:.6f}")
print(f"Insertion Sort  | {i_comp:<11} | {i_swaps:<5} | {i_time:.6f}")

# --- Dynamic Analysis ---
print("\nAnalysis:")
# Find fastest algorithm
times = [("Bubble Sort", b_time), ("Selection Sort", s_time), ("Insertion Sort", i_time)]
fastest = min(times, key=lambda x: x[1])
print(f"- {fastest[0]} was the fastest in this run.")

# Compare swaps
swaps = [("Bubble Sort", b_swaps), ("Selection Sort", s_swaps), ("Insertion Sort", i_swaps)]
least_swaps = min(swaps, key=lambda x: x[1])
print(f"- {least_swaps[0]} performed the least swaps.")

# Compare comparisons
comparisons = [("Bubble Sort", b_comp), ("Selection Sort", s_comp), ("Insertion Sort", i_comp)]
least_comparisons = min(comparisons, key=lambda x: x[1])
print(f"- {least_comparisons[0]} had the fewest comparisons.")

print("- Overall efficiency depends on both time and operations, but the fastest algorithm is recommended for large inputs.")
```
