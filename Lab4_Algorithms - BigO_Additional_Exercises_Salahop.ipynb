```python
import time

# --- Setup ---
student_id = "TUPM-25-4175"
target = int(student_id[-2:])   # last two digits = 12
main_list = list(range(1, 1001))  # 1 to 1000
duplicate_list = [1, 2, 3, 4, 2, 5]

# --- Algorithms ---
def linear_search(arr, target):
    start = time.time()
    for i in range(len(arr)):
        if arr[i] == target:
            end = time.time()
            return i, end - start
    end = time.time()
    return -1, end - start

def binary_search(arr, target):
    start = time.time()
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            end = time.time()
            return mid, end - start
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    end = time.time()
    return -1, end - start

def find_duplicate(arr):
    start = time.time()
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                end = time.time()
                return arr[i], end - start
    end = time.time()
    return None, end - start

# --- Run and Compare ---
lin_index, lin_time = linear_search(main_list, target)
bin_index, bin_time = binary_search(main_list, target)
dup_value, dup_time = find_duplicate(duplicate_list)

print("Algorithm         | Result        | Time (seconds)")
print("-"*50)
print(f"Linear Search     | Index {lin_index:<5} | {lin_time:.6f}")
print(f"Binary Search     | Index {bin_index:<5} | {bin_time:.6f}")
print(f"Duplicate Finder  | Value {dup_value:<5} | {dup_time:.6f}")

# --- Dynamic Analysis ---
print("\nAnalysis:")
if lin_time > bin_time:
    print(f"- Binary search was faster than linear search for target {target}.")
else:
    print(f"- Linear search was surprisingly faster than binary search for target {target} (likely due to small input size).")

print(f"- Duplicate finder identified value {dup_value}, but took {dup_time:.6f} seconds with nested loops.")
print("- This shows how quadratic algorithms scale poorly compared to linear and logarithmic ones.")
```
