```python
# 1. Understanding Algorithms and Pseudocode
# Step 1: Define an algorithm in pseudocode as comments
# Algorithm to find the largest number in a list
# 1. Assume the first number is the largest
# 2. Compare each number with the current largest
# 3. If a number is larger, update the largest
# 4. Return the largest number
# Step 2: Implement it in Python

numbers = [10, 5, 20, 8, 15]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
print("Largest number:", largest)

# TRY IT:
# Finding the smallest number in a list
# 1. Assume the first number is the smallest
# 2. Compare each number with the current smallest
# 3. If a number is smaller, update the smallest
# 4. Return the smallest number
# Implement it on Python:

smallest = numbers[0]
for num in numbers:
    if num < smallest:
        smallest = num
print("Smallest number:", smallest)
```

```python
# 3. Linear Search
# Step 1: Implement linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
        return -1

# Step 2: Test the function
data = [3, 5, 2, 4, 9]
index = linear_search(data, 4)
print("Found at index:", index)

# TRY IT
# Modify the linear search to count the number of comparisons made before finding the target.
def linear_search_count(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons
        return -1, comparisons

data = [3, 5, 2, 4, 9]
index, comps = linear_search_count(data, 4)
print("Found at index:", index, "Comparisons:", comps)
```

```python
# Step 1: Implement binary search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Step 2: Test with a sorted list
sorted_data = [1, 3, 5, 7, 9, 11]
result = binary_search(sorted_data, 7)
print("Found at index:", result)
```

```python
# Try It: Trace how many comparisons are made for different target values
# and relate to O(log n)

def binary_search_trace(arr, target):
    low, high = 0, len(arr) - 1
    comparisons = 0
    while low <= high:
        comparisons += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1, comparisons

# Test with different target values
sorted_data = [1, 3, 5, 7, 9, 11]

for target in [1, 7, 11, 4]:
    result, comps = binary_search_trace(sorted_data, target)
    print("Target:", target, "Result index:", result, "Comparisons:", comps)
```

```python
# Step 1: Implement bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Step 2: Test sorting
to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
print("Sorted list:", to_sort)
```

```python
# Step 1: Implement insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Step 2: Test and measure performance
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
# Try It: Analyze insertion sort's performance when the list is already sorted

def insertion_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    return arr, comparisons, swaps

# Best Case: Already Sorted List
best_case = [1, 2, 3, 4, 5]
sorted_best, comp_best, swap_best = insertion_sort(best_case[:])
print("Best Case:", sorted_best, "Comparisons:", comp_best, "Swaps:", swap_best)

# Average Case: Random List
average_case = [3, 1, 4, 5, 2]
sorted_avg, comp_avg, swap_avg = insertion_sort(average_case[:])
print("Average Case:", sorted_avg, "Comparisons:", comp_avg, "Swaps:", swap_avg)

# Worst Case: Reverse Sorted List
worst_case = [5, 4, 3, 2, 1]
sorted_worst, comp_worst, swap_worst = insertion_sort(worst_case[:])
print("Worst Case:", sorted_worst, "Comparisons:", comp_worst, "Swaps:", swap_worst)
```

```python
# Try It: Experiment with different input sizes and compare the runtime of algorithms using the time module

import time
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

sizes = [100, 200, 300]  # different input sizes

for size in sizes:
    data = random.sample(range(100000), size)

    # Built-in sort (Timsort, O(n log n))
    start = time.time()
    sorted(data)
    builtin_time = time.time() - start

    # Bubble sort (O(n^2))
    start = time.time()
    bubble_sort(data.copy())
    bubble_time = time.time() - start

    # Insertion sort (O(n^2) worst case, O(n) best case)
    start = time.time()
    insertion_sort(data.copy())
    insertion_time = time.time() - start

    # Selection sort (O(n^2))
    start = time.time()
    selection_sort(data.copy())
    selection_time = time.time() - start

    print(f"Input size: {size}")
    print(f"Built-in sort time: {builtin_time:.6f} seconds")
    print(f"Bubble sort time: {bubble_time:.6f} seconds")
    print(f"Insertion sort time: {insertion_time:.6f} seconds")
    print(f"Selection sort time: {selection_time:.6f} seconds")
    print("----")
```

```python
# 10. 
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

print(fib_iter(10))
```

```python
# 11. 
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)

print(fib_rec(10))
```
